La clase Nodo queda:

```python
class Nodo:
    def __init__(self, t):
        self.hijos = [None] * (2 * t + 1)  # Punteros a hijos
        self.llaves = [None] * (2 * t)     # Llaves
        self.hoja = True                  # Indica si el nodo es hoja
        self.n = 0                        # Número de llaves actuales
```

La claseArbolB es:

```python
class ArbolB:
    def __init__(self, gradoMinimo):
        self.t = gradoMinimo
        self.raiz = None

    def bTreeCreate(self):
        if self.raiz is None:
            self.raiz = Nodo(self.t)

    def bTreeSplitChild(self, x, i):
        z = Nodo(self.t)
        y = x.hijos[i]
        z.hoja = y.hoja
        z.n = self.t - 1

        for j in range(self.t - 1):
            z.llaves[j] = y.llaves[j + self.t]
        if not y.hoja:
            for j in range(self.t):
                z.hijos[j] = y.hijos[j + self.t]

        y.n = self.t - 1

        for j in range(x.n, i - 1, -1):
            x.hijos[j + 1] = x.hijos[j]
        x.hijos[i + 1] = z

        for j in range(x.n - 1, i - 2, -1):
            x.llaves[j + 1] = x.llaves[j]
        x.llaves[i] = y.llaves[self.t - 1]
        x.n += 1

    def bTreeInsertNonFull(self, x, k):
        i = x.n - 1
        if x.hoja:
            while i >= 0 and k < x.llaves[i]:
                x.llaves[i + 1] = x.llaves[i]
                i -= 1
            x.llaves[i + 1] = k
            x.n += 1
        else:
            while i >= 0 and k < x.llaves[i]:
                i -= 1
            i += 1
            if x.hijos[i].n == 2 * self.t - 1:
                self.bTreeSplitChild(x, i)
                if k > x.llaves[i]:
                    i += 1
            self.bTreeInsertNonFull(x.hijos[i], k)

    def bTreeInsert(self, k):
        if self.raiz.n == 2 * self.t - 1:
            s = Nodo(self.t)
            s.hoja = False
            s.hijos[0] = self.raiz
            self.raiz = s
            self.bTreeSplitChild(s, 0)
            self.bTreeInsertNonFull(s, k)
        else:
            self.bTreeInsertNonFull(self.raiz, k)

    def imprimirArbol(self, nodo, nivel=0):
        if nodo is not None:
            print("Nivel", nivel, ":", nodo.llaves[:nodo.n])
            if not nodo.hoja:
                for hijo in nodo.hijos[:nodo.n + 1]:
                    self.imprimirArbol(hijo, nivel + 1)
```

```python
# Clase controladora
class Controladora:
    def main(self):
        secuencia = "BTHMOCZGLENPRDJQFWX"
        BT = ArbolB(2)
        BT.bTreeCreate()

        print("Creando e insertando los elementos en el árbol B:")
        for caracter in secuencia:
            print(f"\nInsertando: {caracter}")
            BT.bTreeInsert(caracter)
            print("Árbol actual:")
            BT.imprimirArbol(BT.raiz)
```

```python
if __name__ == "__main__":
    Controladora().main()
```
